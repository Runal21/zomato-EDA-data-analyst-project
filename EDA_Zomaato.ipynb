### I'll clean the dataset using these steps:
#### ✅ Fill missing values
#### ✅ Convert data types
#### ✅ Handle inconsistencies
#### ✅ Remove duplicates

```python
# Step 1: Fill Missing Values
# Fill 'zipcode' with 'Unknown' (since most values are missing)
df1['zipcode'].fillna('Unknown', inplace=True)
# Fill 'timings' with 'Not Available'
df1['timings'].fillna('Not Available', inplace=True)

# Step 2: Convert Data Types
df1['aggregate_rating'] = df1['aggregate_rating'].astype(float)

# Step 3: Handle Inconsistencies
# Convert -1 values in 'delivery' and 'takeaway' to 0 (assuming -1 means 'not available')
df1['delivery'].replace(-1, 0, inplace=True)
df1['takeaway'].replace(-1, 0, inplace=True)

# Step 4: Remove Duplicates
df1.drop_duplicates(inplace=True)

# Verify cleaning results
df1.info(), df1.head()
```

### I'll proceed with Exploratory Data Analysis (EDA) by:
#### ✅ Generating basic statistics
#### ✅ Identifying top cities with the most restaurants
#### ✅ Analyzing popular cuisines
#### ✅ Visualizing price range distribution
#### ✅ Examining rating trends

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set_style("whitegrid")

# Step 1: Basic Statistics
summary_stats = df1.describe()

# Step 2: Top 10 Cities with Most Restaurants
top_cities = df1['city'].value_counts().head(10)

# Step 3: Most Popular Cuisines
# Splitting the cuisines and counting occurrences
from collections import Counter
cuisine_list = df1['cuisines'].str.split(', ').explode()
top_cuisines = pd.Series(Counter(cuisine_list)).sort_values(ascending=False).head(10)

# Step 4: Price Range Distribution
price_range_counts = df1['price_range'].value_counts().sort_index()

# Step 5: Ratings Distribution
rating_distribution = df1['aggregate_rating'].value_counts().sort_index()

# Plot: Top 10 Cities with Most Restaurants
plt.figure(figsize=(10, 5))
sns.barplot(x=top_cities.index, y=top_cities.values, palette='Blues_r')
plt.xticks(rotation=45)
plt.title('Top 10 Cities with Most Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.show()

# Plot: Top 10 Cuisines
plt.figure(figsize=(10, 5))
sns.barplot(x=top_cuisines.index, y=top_cuisines.values, palette='Greens_r')
plt.xticks(rotation=45)
plt.title('Top 10 Most Popular Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.show()

# Plot: Price Range Distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=price_range_counts.index, y=price_range_counts.values, palette='Oranges_r')
plt.title('Restaurant Distribution by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.show()

# Plot: Ratings Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df1['aggregate_rating'], bins=10, kde=True, color='purple')
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

summary_stats
```
