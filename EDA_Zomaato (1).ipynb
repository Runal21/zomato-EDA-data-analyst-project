# Data Cleaning Findings:

**Next Steps:**

✅ Drop Duplicates (since they are redundant).


✅ Drop zipcode (too many missing values, likely not useful).


✅ Fill timings with "Not Available".

✅ Check Data Types after cleanup.

```python
# Drop duplicate rows
df_cleaned = df1.drop_duplicates()

# Drop the 'zipcode' column due to excessive missing values
df_cleaned = df_cleaned.drop(columns=["zipcode"])

# Fill missing values in 'timings' with "Not Available"
df_cleaned["timings"] = df_cleaned["timings"].fillna("Not Available")

# Verify changes
df_cleaned.isnull().sum(), df_cleaned.duplicated().sum()
```

```python
# Check data types of all columns
df_cleaned.dtypes
```

**Fix Data Types**

```python
# Convert numeric columns
df_cleaned["aggregate_rating"] = pd.to_numeric(df_cleaned["aggregate_rating"], errors="coerce")
df_cleaned["votes"] = pd.to_numeric(df_cleaned["votes"], errors="coerce")
df_cleaned["average_cost_for_two"] = pd.to_numeric(df_cleaned["average_cost_for_two"], errors="coerce")

# Convert categorical columns
df_cleaned["rating_text"] = df_cleaned["rating_text"].astype("category")
df_cleaned["city"] = df_cleaned["city"].astype("category")
df_cleaned["establishment"] = df_cleaned["establishment"].astype("category")
```

**Standardize Text Formatting**

```python
# Remove extra spaces and special characters
df_cleaned["cuisines"] = df_cleaned["cuisines"].str.strip().str.title()
df_cleaned["highlights"] = df_cleaned["highlights"].str.replace(r"[^\w\s,]", "", regex=True).str.title()
df_cleaned["timings"] = df_cleaned["timings"].str.strip().replace("", "Not Available")
```

**Handle Outliers (IQR Method)**

```python
Q1 = df_cleaned["average_cost_for_two"].quantile(0.25)
Q3 = df_cleaned["average_cost_for_two"].quantile(0.75)
IQR = Q3 - Q1

# Define limits
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Cap outliers
df_cleaned["average_cost_for_two"] = df_cleaned["average_cost_for_two"].clip(lower=lower_bound, upper=upper_bound)
```

**Normalize List Values (e.g., Establishments, Highlights)**

```python
import ast

# Function to safely convert string lists to actual lists
def safe_convert(value):
    try:
        return ast.literal_eval(value) if isinstance(value, str) else value
    except (SyntaxError, ValueError):
        return [value] if isinstance(value, str) else []

# Apply function to problematic columns
df_cleaned["establishment"] = df_cleaned["establishment"].apply(safe_convert)
df_cleaned["highlights"] = df_cleaned["highlights"].apply(safe_convert)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure numeric columns are properly formatted
df_cleaned["aggregate_rating"] = pd.to_numeric(df_cleaned["aggregate_rating"], errors="coerce")
df_cleaned["votes"] = pd.to_numeric(df_cleaned["votes"], errors="coerce")
df_cleaned["average_cost_for_two"] = pd.to_numeric(df_cleaned["average_cost_for_two"], errors="coerce")

# Ensure 'name' column is a string
df_cleaned["name"] = df_cleaned["name"].astype(str)

# Count the number of outlets for each restaurant chain
restaurant_chains = df_cleaned["name"].value_counts()

# Top restaurant chains by number of outlets
top_chains_outlets = restaurant_chains.head(10)

# Top restaurant chains by average rating
top_chains_ratings = df_cleaned.groupby("name")["aggregate_rating"].mean().sort_values(ascending=False).head(10)

# Plot: Top Restaurant Chains by Number of Outlets
plt.figure(figsize=(10, 5))
top_chains_outlets.sort_values().plot(kind='barh', color=sns.color_palette("husl", len(top_chains_outlets)))

plt.xlabel("Number of Outlets")
plt.ylabel("Restaurant Chain")
plt.title("Top 10 Restaurant Chains by Number of Outlets")
plt.show()

# Plot: Top Restaurant Chains by Average Rating
plt.figure(figsize=(10, 5))
top_chains_ratings.sort_values().plot(kind='barh', color=sns.color_palette("coolwarm", len(top_chains_ratings)))

plt.xlabel("Average Rating")
plt.ylabel("Restaurant Chain")
plt.title("Top 10 Restaurant Chains by Average Rating")
plt.xlim(4.0, 5.0)  # Focus on high ratings
plt.show()
```

Most restaurants focus on customer convenience, offering takeaway, seating comfort, and flexible payment options. Family-friendly options (No Alcohol Available) are prioritized in many places.

```python
# Number of restaurants per highlight
# Drop missing values in 'highlights' and 'aggregate_rating'
df_cleaned = df_cleaned.dropna(subset=["highlights", "aggregate_rating"])

# Explode highlights into individual features
highlight_ratings = df_cleaned.copy()
highlight_ratings["highlights"] = highlight_ratings["highlights"].str.split(", ")
highlight_ratings = highlight_ratings.explode("highlights")

# Group by highlight and calculate average rating
highlight_ratings = highlight_ratings.groupby("highlights")["aggregate_rating"].mean().sort_values(ascending=False).head(10)

# Plot bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=highlight_ratings.values, y=highlight_ratings.index, palette="viridis")

# Labels and title
plt.xlabel("Average Rating", fontsize=14, fontweight="bold")
plt.ylabel("Highlight Feature", fontsize=14, fontweight="bold")
plt.title("Top 10 Highest Rated Highlights", fontsize=16, fontweight="bold")

# Show plot
plt.show()

```

💡**Insight:**

Restaurants that provide a unique dining experience—whether through music, ambiance, exclusivity, or inclusivity (vegan, pet-friendly)—receive significantly higher ratings from customers.

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Ensure 'highlights' column is string and drop NaN values
df_cleaned["highlights"] = df_cleaned["highlights"].astype(str)
df_cleaned["highlights"] = df_cleaned["highlights"].replace("nan", "")  # Replace 'nan' strings

# Generate word cloud
highlight_text = " ".join(df_cleaned["highlights"].dropna().str.replace(", ", " "))  # Remove commas
wordcloud = WordCloud(width=800, height=400, background_color="black", colormap="coolwarm").generate(highlight_text)

# Plot the word cloud
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Restaurant Highlights", fontsize=16, fontweight="bold")

# Show the plot
plt.show()
```
